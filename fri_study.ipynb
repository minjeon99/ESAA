## **DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환**

### **넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기**
- DataFrame은 칼럼명을 가지고 있어 리스트, ndarray보다 편하게 데이터 핸들링 가능
- DataFrame으로 변환 시 칼럼명 지정
- DataFrame 객체
  - 생성 인자 data: 리스트, 딕셔너리, ndarray 입력
  - 생성 인자 columns: 칼럼명 리스트 입력

````python
# 1차원

import numpy as np

col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
# 리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)
# 넘파이 ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFram:\n', df_array1)
````

````python
# 2차원 (2행 3열 형태)

# 3개의 칼럼명 필요
col_name2 = ['col1', 'col2', 'col3']

# 2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환
list2 = [[1,2,3],
         [11,12,13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape)
df_list2 = pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)
df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
````

- 딕셔너리 -> DataFrame
  - 키(key): 칼럼명
  - 값(value): 키에 해당하는 칼럼 데이터로 변환

    => 키: 문자열, 값: 리스트(or ndarray) 형태로 딕셔너리 구성

````python
# Key는 문자열 칼럼명으로 매핑, Value는 리스트 형(또는 ndarray) 칼럼 데이터로 매핑
dict = {'col1':[1,11], 'col2':[2,22], 'col3':[3,33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
````

### **DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기**

- 많은 머신러닝 패키지가 기본 데이터 형으로 넘파이 ndarray를 사용
  -> 데이터 핸들링에 DataFrame 이용하더라도 다시 ndarray로 변환하는 경우 빈번하게 발생
- DataFrame 객체의 values를 이용해 쉽게 변환 가능

````python
# DataFrame을 ndarray로 변환
array3 = df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape)
print(array3)
````

- 리스트로 변환: tolist() 호출
- 딕셔너리로 변환: to_dict() 호출
  - 인자로 'list' 입력하면 딕셔너리 값이 리스트형으로 반환

````python
# DataFrame을 리스트로 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

# DataFrame을 딕셔너리로 변환
dict3 = df_dict.to_dict('list')
print('\n df_dict.values.to_dict() 타입:', type(dict3))
print(dict3)
````

## **DataFrame 데이터 삭제**
- drop() 메서드 원형
```
  DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
```



**axis**: 칼럼 또는 행 드롭 지정
- axis=1 -> 칼럼 드롭: 주로 칼럼 드롭 이용. 기존 칼럼 값 가공해 새 칼럼 만들고 삭제하는 경우 다수.
- axis=0 -> 로우 드롭: 주로 이상치 데이터 삭제하는 경우에 이용
  - 특정 로우를 가리키는 것=인덱스 -> 자동으로 labels에 오는 값을 인덱스로 간주

````python
# Age_0 칼럼 삭제
titanic_drop_df = titanic_df.drop('Age_0', axis=1)
titanic_drop_df.head(3)
````

**inplace**
- **inplace = False**(default): 자기 자신의 DataFrame 데이터는 **삭제하지 않고**, 삭제된 결과 DataFrame을 반환

````python
titanic_df.head(3)
# 'Age_0' 칼럼 여전히 존재
````

- 여러 개의 칼럼 삭제: 리스트 형태로 칼럼 명 입력해 labels 파라미터로 입력

````python
drop_result = titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis=1, inplace=True)
print('inplace=True로 drop 후 반환된 값:', drop_result)
titanic_df.head(3)
````

- drop() 시 inplace=True -> 반환 값이 **None**
  ```
  titanic_df = titanic_df.drop(['Age_0', 'Age_by_10'])
  ```
  -> titanic_df 객체 변수가 아예 None이 됨


````python
# 로우 삭제
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', 15)
print('#### before axis 0 drop ####')
print(titanic_df.head(3))
titanic_df.drop([0,1,2], axis=0, inplace=True)

print('#### after axis 0 drop ####')
print(titanic_df.head(3))
````

- axis: DataFrame의 로우를 삭제할 때는 axis=0, 칼럼을 삭제할 때는 axis=1
- 원본 DataFrame은 유지하고 드롭된 DataFrame을 새롭게 객체 변수로 받고 싶다면 inplace=False(디폴트 값이 False)
- 원본 DataFrame에 드롭된 결과를 적용할 경우 inplace=Trure
- 원본 DataFrame에서 드롭된 DataFrame을 다시 원본 DataFrame 객체 변수로 할당하면 원본 DataFrame에서 드롭된 결과를 적용할 경우와 같음

  ```
  titanic_df = titanic_df.drop('Age_0', axis=1, inplace=False)
  ```

## **데이터 셀렉션 및 필터링**

- 넘파이와 유사한 부분도, 다른 부분도 있음

### **DataFrame의 [ ] 연산자**
- 넘파이의 '[ ]'

  - 칼럼 명 문자 또는 인덱스로 변환 가능한 표현식
- DataFrame 뒤에 있는 [ ]
  
  : 칼럼만 지정할 수 있는 '칼럼 지정 연산자'
  - 여러 개의 칼럼에서 데이터 추출 -> ['칼럼1', '칼럼2']
  - titanic_df[0] -> 오류! (0은 칼럼 명이 아니기 때문)

````python
print('단일 칼럼 데이터 추출:\n', titanic_df['Pclass'].head(3))
print('\n여러 칼럼의 데이터 추출:\n', titanic_df[['Survived', 'Pclass']].head(3))
print('[ ] 안에 숫자 index는 KeyError 오류 발생:\n', titanic_df[0])
````

- '[ ]' 내 숫자 값 입력 -> 오류
- 판다스의 인덱스 형태로 변환 가능한 표현식은 입력 가능 e.g. 슬라이싱 (O)

- DataFrame[0:2]와 같은 슬라이싱 연산으로 데이터 추출하는 방법은 사용하지 않는 것 권장

### **DataFrame ix[ ] 연산자**
- 넘파이 ndarray의 [ ] 연산자와 유사한 기능 제공하기 위해 제공
- ix[0, 'Pclass']와 같이 행 위치 지정으로 인덱스값 0, 열 위치 지정으로 칼럼 명 'Pclass' 입력해 원하는 위치의 데이터 추출 가능
  - 행 위치 지정: DataFrame의 인덱스값 입력
  - 열 위치 지정
    - 칼럼 명 지정 -> 칼럼 명칭(label) 기반 인덱싱
    - 칼럼의 위치 값(ix[0, 2]) 지정 -> 칼럼 위치(position) 기반 인덱싱
- ix[ ]는 현재 판다스에서 사라짐

````python
print('칼럼 위치 기반 인덱싱 데이터 추출:', titanic_df.ix[0,2])
print('칼럼 명 기반 인덱싱 데이터 추출:', titanic_df.ix[0,'Pclass'])
````

- 넘파이 ndarrya의 [ ]와 동일하게 단일 지정, 슬라이싱, 불린 인덱싱, 팬시 인덱싱 모두 가능

````python
data = {'Name': ['Chulmin', 'Eunkyung', 'Jinwoong', 'Soobeom'],
        'Year': [2011,2016,2015,2015],
        'Gender': ['Male', 'Female', 'Male', 'Male']}
data_df = pd.DataFrame(data, index=['one', 'two', 'three', 'four'])
data_df
````

### **명칭 기반 인덱싱과 위치 기반 인덱싱의 구분**
- 명칭 기반 인덱싱: 칼럼의 명칭을 기반으로 위치 지정
  - '칼럼 명' 같이 명칭으로 열 위치 지정하는 방식
- 위치 기반 인덱싱: 0을 출발점으로 하는 가로축, 세로축 좌표 기반의 행과 열 위치를 기반으로 데이터 지정

````python
# data_df를 reset_index()로 새로운 숫자형 인덱스를 생성
data_df_reset = data_df.reset_index()
data_df_reset = data_df_reset.rename(columns={'index':'old_index'})

# 인덱스값에 1을 더해서 1부터 시작하는 새로운 인덱스값 생성
data_df_reset.index = data_df_reset.index+1
data_df_reset
````

- 인덱스가 integer형인 경우 ix[0,1]는 DataFrame의 인덱스값 사용

  (행 위치 값으로 위치 기반 인덱스값 사용 X)

  => 인덱스값이 0인 key는 찾을 수 없음

````python
data_df_reset.ix[1,1]
````

### **DataFrame iloc[ ] 연산자**
- iloc[ ]: 위치 기반 인덱싱만 허용
  - 행과 열 값으로 integer / integer형의 슬라이싱, 팬시 리스트 값 입력
  - 불린 인덱싱은 제공하지 않음

````python
data_df.iloc[0,0]
````

````python
# 다음 코드는 오류를 발생합니다.
data_df.iloc[0,'Name']
````

````python
data_df_reset.iloc[0,1]
````

- ix[0,1]은 행 위치에 인덱스값 0을 찾지 못해 오류 발생했지만
- iloc[0,1]은 위치 기반 인덱싱 -> 첫 번째 행(0)의 두 번째 열(1)의 위치에 있는 값 반환

### **DataFrame loc[ ] 연산자**
- loc[ ]: 명칭 기반으로 데이터 추출
  - 행 위치: DataFrame index 값
  - 열 위치: 칼럼 명

````python
data_df.loc['one','Name']
````

- index가 숫자 형일 수 있기 때문에 항상 문자열을 입력하는 것은 아님

## **apply lambda 식으로 데이터 가공**
- DataFrame이나 Series의 레코드별로 데이터 가공
- 칼럼에 일괄적으로 데이터 가공하는 것이 속도 면에서 더 빠르나, 복잡한 데이터 가공 필요한 경우 apply lambda 이용



````python
def get_square(a):
  return a**2

print('3의 제곱은:', get_square(3))
````

- lambda 식: 파이썬에서 함수형 프로그래밍을 지원하기 위해 만들어짐
  - 함수 선언과 함수 내 처리를 한 줄의 식으로 쉽게 변환
  - ':'로 입력 인자와 반환될 입력 인자의 계산식 분리
    - 입력 인자: ':'의 왼쪽 (x)
    - 반환 값: ':'의 오른쪽. 입력 인자의 계산식
  - 여러 개의 입력 인자 -> map() 함수와 결합해서 사용

````python
# get_square(a) 함수 -> lambda 식으로 변환하기

lambda_square = lambda x : x ** 2
print('3의 제곱은:', lambda_square(3))
````

````python
a = [1,2,3]
squares = map(lambda x: x**2, a)
list(squares)
````

**판다스 DataFrame에 lambda 식 적용하기**

````python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name', 'Name_len']].head(3)
````

````python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x:'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
````

- lambda 식의 if else
  - if 절: if 식보다 반환 값을 먼저 기술

    (lambda 식 ':' 기호의 오른편에 반환 값이 있어야 함)
  - else: else 식이 먼저 나오고, 반환 값이 나중
  - else if는 지원 X

    => else 절을 ()로 내포해 () 내에서 다시 if else 적용

````python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
````

- else if가 많이 나오는 경우 or switch case 문의 경우 **별도 함수 생성** 권장

````python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성

def get_category(age):
  cat = ''
  if age <= 5: cat='Baby'
  elif age <= 12: cat='Child'
  elif age <= 18: cat='Teenager'
  elif age <= 25: cat='Student'
  elif age <= 35: cat='Young Adult'
  elif age <= 60: cat='Adult'
  else: cat='Elderly'

  return cat
````

````python
# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age','Age_cat']].head()
````
